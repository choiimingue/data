## An introduction to seaborn

### Seaborn 소개
- seaborn은 python에서 통계 그래픽을 만드는 라이브러리
- matplotlib을 기반으로 하며, pandas 데이터 구조와 밀접하게 통합됨
- plotting 기능은 전체 데이터셋을 포함하는 DataFrame, Arrary에서 작동함 
- 내부적으로 필요한 [Semantic mapping][1] 및 통계 집계를 수행하여 플롯을 생성함
- 데이터셋 지향적인 선언형 API는 플롯을 어떻게 그려야 하는지 세세하게 신경쓰지 않고 plot의 다른 요소들이 무엇인지에 집중할 수 있게 도와줌



**Semantic mapping**
- 고차원 데이터를 차원 축소하는 것
- 데이터 단순화를 통해 새로운 것을 발굴하거나 시스템을 구축하기 위함
[1]:https://en.wikipedia.org/wiki/Semantic_mapping_(statistics)

**`relplot()`**
- [`relplot()`][8]
- tip Dataset의 5개 변수 사이의 관계를 보여줌
- 플롯에서 변수의 이름과 역할을 지정하는 방법이 matplotlib에서 일일이 요소들을 지정해주는 것에 비해 매우 단순함
- Seaborn이 내부에서 DataFrame의 값을 matplotlib이 이해하는 argument로 알아서 변환하여 전달하는 방식

**선언적 접근방식**
- matplotlib을 제어하는 방법에 대한 세부 요소에 집중하기 보다는 시각화를 통해 답변하려는 의문, 질문(본질)에 집중할 수 있게 해줌

[8]:https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

### 통계 그래픽을 위한 high-level API
- 데이터를 시각화하는 보편적인 최선의 방법은 없음
- 다양한 질문과 의문, 목적에 대응하기 위해 다양한 플롯을 사용함
- seaborn은 데이터셋 지향 API를 지향하여 다양한 시각적 표현을 쉽게 전환할 수 있도록 도와줌

**`relplot()`**
- 다양한 통계적 관계(relationship)를 시각화하도록 설계된 플롯
- 산점도를 나타내기에 효과적이지만, 하나의 변수가 시간 측정을 나타내는 관계는 선으로 더 잘 표현됨
- [`relplot()`][8]은 쉽게 대체 표현으로 전환할 수 있는 간편한 `kind` parameter를 포함하고 있음

[8]:https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
dots = sns.load_dataset("dots")
sns.relplot(
    data=dots, kind="line",
    x="time", y="firing_rate", col="align",
    hue="choice", size="coherence", style="choice",
    facet_kws=dict(sharex=False),
)

dots
```

- `size` 및 `style` parameter는 산점도(scatter)와 선(line) 플롯 모두에서 사용되지만
- 두 시각화에 미치는 영향은 다른 방식으로 나타남
- 산점도(scatter) 플롯에서는 marker 영역과 symbol을 변경하고
- 선(line) 플롯에서는 line width와 dashing을 변경함
- seaborn을 사용할 때에는 이러한 세부 사항을 일일이 고려할 필요가 없으므로 플롯의 전체 구조와 전달하려는 정보에 집중할 수 있음

### 통계적 추정
- 한 변수의 평균값을 다른 변수의 함수로 확인하는 작업을 하기도 함
- 많은 seaborn 함수는 통계적 추정을 자동으로 수행함

```python
fmri = sns.load_dataset("fmri")
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", col="region",
    hue="event", style="event",
)

fmri
```

**seaborn의 통계적 추정**
- 통계 값이 추정되면 seaborn은 부트스트래핑을 사용하여 신뢰구간을 계산함
- 계산된 신뢰구간을 토대로 추정의 불확실성을 나타내는 오차 막대를 그려줌
- seaborn의 통계적 추정은 단순 기술 통계의 한계를 넘어섬
- 예를 들어, `lmplot()`을 사용하면 선형 회귀 모델(및 불확실성)을 포함한 산점도를 그릴 수 있음

```python
sns.lmplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker")
```

### 분포 표현
[**`displot()`**][1]
- 통계 분석에는 데이터셋의 변수 분포에 대한 지식이 필요함
- seaborn 함수 `displot()`은 분포를 시각화하는 여러 접근 방식을 지원함
- histogram과 같은 고전적인 기술, kde(커널 밀도 추정)과 같은 계산 집약적 접근방식이 포함됨

[1]:https://seaborn.pydata.org/generated/seaborn.displot.html#seaborn.displot

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
```

**`ecdf plot`**
- 데이터의 경험적 누적 분포 플롯
- seaborn은 강력하지만 덜 친숙한 기술도 지원함

```python
sns.displot(data=tips, kind="ecdf", x="total_bill", col="time", hue="smoker", rug=True)
```

### 범주형 데이터 플롯
[**`catplot()`**][1]
- seaborn의 범주형 데이터 시각화
- 다양한 level의 세분화를 제공

[1]:https://seaborn.pydata.org/generated/seaborn.catplot.html#seaborn.catplot

**`swarm plot`**
- `catplot(kind="swarm")`
- 가장 미세한 level
- 군집도를 그려서 모든 관측값을 볼 수 있고, 범주 축을 따라 점의 위치를 조정함

```python
sns.catplot(data=tips, kind="swarm", x="day", y="total_bill", hue="smoker")
```

**`violin plot`**
- `catplot(kind="violin")`
- 커널 밀도 추정을 사용
- 포인트가 샘플링되는 기본 분포를 나타낼 수도 있음

```python
sns.catplot(data=tips, kind="violin", x="day", y="total_bill", hue="smoker", split=True)
```

**`bar plot`**
- `catplot(kind="bar")`
- 각 중첩 범주 내에서 평균값과 신뢰 구간만 표시할 수 있음

```python
sns.catplot(data=tips, kind="bar", x="day", y="total_bill", hue="smoker")
```
